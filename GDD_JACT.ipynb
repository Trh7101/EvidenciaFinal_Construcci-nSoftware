## _Game Design_

---

### **Summary**

A game that allows users to create custom racing-track courses by designing the track, environment, and item & obstacle layout; configuring gamplay rules and the overall look of the design based on switchable themes. Users can share or play levels of their own or created by other users.

### **Gameplay**

#### Level Design
Users should be able to design levels without any programming or game development experience via a simple drag and drop interface. Users can choose from an array of pieces and drag them onto a grid to design the main level. Examples of these include track segments, items, obstacles and scenery. Users can then configure additional settings through a menu such as the theme of their course (spring, summer, fall, winter), speed of the cars, time limit, etc. Finally, they can preview their course gameplay and publish it so other users can play their levels.

#### Level Selection
Users can browse levels created by other users and play them. Levels can optionally be organized in collections, as to create a complete game or group similar courses. Levels and collections can be searched and filtered using full text search.

#### Level Gameplay
In order to finish a game, users must complete a number of laps (set by the level creator). Additionally, levels may have a time limit. Two gamplay modes will be implemented (level creators can choose which one to use):
* Players cannot move outside of the track and must go through all of the track in order for a lap to be counted.
* Players can move freely but must pass through a series of checkpoints in order for the lap to be counted. This can increase difficulty by making it easy for the player to go off track and forcing them to go back.

Obstacles placed in the track will make it harder for a player to complete a lap. Whenever a player hits an obstacle, an event may be triggered depending on the item type, for instance, on collision with an oil puddle they will be unable to move for a few seconds, falling on a lake will make them loose a life. When a player looses a ceirtain amount of lives (set by the level creator) the game will be over.

Items placed in the track will make it easier for a player to complete a lap. Examples of items include speed ramps and extra lives.

Users should be aware of the placement of obstacles and items in order to improve their time.

The main challenge in the gameplay should be completing levels in as little time as possible. After each game, the time in which the level was completed will be posted to a leaderboard, this way users can compete against other users.

### **Mindset**
The main goal of this game is to teach the fundamentals of game design and for anyone to be able to design their own levels and tracks. And in doing so, we expect that the players when making their tracks can feel both the struggle to make it functional and interesting, and also the acomplishment in being able to play it.

In other hand when playing with other track designs, we expect that the players try as many levels while they try their best to acomplish record laps. That way they could feel the eager to compete to get the best times of as many levels as they play.
